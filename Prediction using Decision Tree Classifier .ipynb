# Prediction using decision tree classifier

# Import libraries and load datasets.
Perform Exploratory data analysis

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split      
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
db=pd.read_csv(r"E:\Let's grow more\Iris.csv")  #Load data set
```

```python
db.head()
```

```python
db.isna().sum()
```

```python
data=pd.DataFrame(db)
```

```python
x=db[['Sepal_length','sepal_width','Petal_Length','Petal_width']]
y=db['Species']
```

```python
x
```

```python
y
```

# Divide dataset into training and testing

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.25,random_state=0)
```

# Decision Tree model

Models for training, Testing, Criteria=Gini and criteria for entrophy

```python
model = DecisionTreeClassifier()

# Train Decision Tree Classifer
model = model.fit(x_train,y_train)

#Predict the response for test dataset
y_pred_train = model.predict(x_train)
```

```python
y_pred_train
```

```python
model2 = DecisionTreeClassifier(max_depth=5,random_state=0)

# Train Decision Tree Classifer
model2 = model2.fit(x_train,y_train)

#Predict the response for test dataset
y_pred_test = model2.predict(x_test)
y_pred_test
```

```python
model3 = DecisionTreeClassifier(criterion="entropy", max_depth=5)

# Train Decision Tree Classifer
model3 = model3.fit(x_train,y_train)


y_pred_entrophy = model3.predict(x_test)
y_pred_entrophy
```

```python
model4 = DecisionTreeClassifier(criterion="gini", max_depth=5)

# Train Decision Tree Classifer
model4 = model4.fit(x_train,y_train)


y_pred_gini = model4.predict(x_test)
y_pred_gini
```

# Priniting Accuracy for all models 

# Accuracy is greater than 0.97 or 97 % so we successfully predicted desired output class

```python
# Model Accuracy
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

print("Accuracy of training :",metrics.accuracy_score(y_train, y_pred_train))

print("Accuracy of testing :",metrics.accuracy_score(y_test, y_pred_test))

print("Accuracy of model with criteria is entrophy :",metrics.accuracy_score(y_test, y_pred_entrophy))

print("Accuracy of model criteria is gini :",metrics.accuracy_score(y_test, y_pred_gini))
```

# Printing Confusion matrix for all models

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix for training : \n\n",confusion_matrix(y_train,y_pred_train),"\n\n")

print("Confusion matrix for testing : \n\n",confusion_matrix(y_test,y_pred_test),"\n\n")

print("Confusion matrix for criteria = entrophy : \n\n",confusion_matrix(y_test,y_pred_entrophy),"\n\n")

print("Confusion matrix for criteria = gini : \n\n",confusion_matrix(y_test,y_pred_gini),"\n\n")
```

# Printing Report for all models

```python
print("Report for model training : \n\n",classification_report(y_train,y_pred_train))
```

```python
print("Report for model testing : \n\n",classification_report(y_test,y_pred_test))
```

```python
print("Report for model entrophy : \n\n",classification_report(y_test,y_pred_gini))
```

```python
print("Report for model gini : \n\n",classification_report(y_test,y_pred_entrophy))
```
